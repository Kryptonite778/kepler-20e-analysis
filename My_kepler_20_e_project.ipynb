##*Introduction*
Kepler-20 e is a terrestrial exoplanet that orbits a (G-type star) Sun.
It is notable for being one of the first Earth-sized exoplanets found orbiting a Sun-like star.
Studying such planets helps us understand planetary systems beyond our solar system,
their orbital properties, and potential similarities to Earth.

###3.1 River Plot
